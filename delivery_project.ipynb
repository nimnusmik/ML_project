### Timetaken 데이터 int 화 

```python
train['Time_taken(min)']
```

```python
train["Time_taken(min)"] = train["Time_taken(min)"].str.replace('(', '',)
train["Time_taken(min)"] = train["Time_taken(min)"].str.replace(')', '')
train["Time_taken(min)"] = train["Time_taken(min)"].str.replace('min', '')
train["Time_taken(min)"]
```

```python
train["Time_taken(min)"] = train["Time_taken(min)"].apply(pd.to_numeric)
train["Time_taken(min)"]
```

```python
train.info()
```

```python
train.head()
```

### 날짜데이터

```python
train['Order_date_new'] = pd.to_datetime(train['Order_Date'])
train['Order_date_new']
```

```python
train['order_year'] = train['Order_date_new'].dt.year
train['order_month'] = train['Order_date_new'].dt.month
train['order_day'] = train['Order_date_new'].dt.day
train['order_week'] = train['Order_date_new'].dt.day_name()
```

```python
train.head()
```

```python
train = train.drop('Order_Date',axis=1)
train.head()
```

```python
train.info()
```

### picked time

```python
from datetime import datetime

train['Time_Order_picked'] = datetime(train['Time_Order_picked'],format='%H:%M:%S')
train['Time_Orderd'] = datetime(train['Time_Orderd'],format='%H:%M:%S')
```

```python
train['diff_picked'] = train['Time_Order_picked'] - train['Time_Orderd'] 
```

### 라벨인코딩

문자형 변수를 숫자형 범수로

```python
object_features = train[['Weatherconditions','Road_traffic_density','Type_of_order','Type_of_vehicle','Festival','City']]
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
for f in object_features:
    labels = []
    le = LabelEncoder()
    items = list(train[f])
    le.fit(items)
    labels = le.transform(items)
    train[f] = labels
    
```
