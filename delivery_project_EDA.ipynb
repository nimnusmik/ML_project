### 결측치 확인 및 대치 

```python
train.isnull().sum()
```

```python
train.isna().sum()/train.shape[0] 
```

```python
train.columns
```

```python
# ID 컬럼 2개 제거 
train = train.drop(['ID','Delivery_person_ID'], axis = 1)
train.head()
```

```python
# 결측치 있는 열 확인
train.columns[train.isnull().any()]
```

```python
# Age, Rating ; 평균값으로 대체
train["Delivery_person_Age"] = train["Delivery_person_Age"].apply(pd.to_numeric)
train["Delivery_person_Age"]
```

```python
train['Delivery_person_Age'].fillna(train['Delivery_person_Age'].mean(), inplace=True)
```

```python
train[train.isnull().any(axis=1)]
```

```python
train["Delivery_person_Ratings"] = train["Delivery_person_Ratings"].apply(pd.to_numeric)
train["Delivery_person_Ratings"]
```

```python
train['Delivery_person_Ratings'].fillna(train['Delivery_person_Ratings'].mean(), inplace=True)
```

```python
# Age, Rating 평균치 넣어서 결측값 없앤 것 확인
train.isnull().sum()
```

```python
# 나머지는 NaN 있는 행 전부 drop
train.dropna()
```

```python
train.columns[train.isnull().any()]
```

```python
train.dropna(subset=['Time_Orderd', 'Weatherconditions', 'Road_traffic_density',
       'multiple_deliveries', 'Festival', 'City'], inplace=True)
```

```python
train.isnull().sum()
```

### Timetaken 데이터 int 화 

```python
train['Time_taken(min)']
```

```python
train["Time_taken(min)"] = train["Time_taken(min)"].str.replace('(', '',)
train["Time_taken(min)"] = train["Time_taken(min)"].str.replace(')', '')
train["Time_taken(min)"] = train["Time_taken(min)"].str.replace('min', '')
train["Time_taken(min)"]
```

```python
train["Time_taken(min)"] = train["Time_taken(min)"].apply(pd.to_numeric)
train["Time_taken(min)"]
```

```python
train.info()
```

```python
train.head()
```

### 날짜데이터

```python
train['Order_date_new'] = pd.to_datetime(train['Order_Date'])
train['Order_date_new']
```

```python
train['order_year'] = train['Order_date_new'].dt.year
train['order_month'] = train['Order_date_new'].dt.month
train['order_day'] = train['Order_date_new'].dt.day
train['order_week'] = train['Order_date_new'].dt.day_name()
```

```python
train.head()
```

```python
train = train.drop('Order_Date',axis=1)
train.head()
```

```python
train.info()
```

### picked time

```python
train['Time_Order_picked']
```

```python
train['diff_picked'] = train['Time_Order_picked'] - train['Time_Orderd']
train['diff_picked']
```

```python
train.head()
```

```python
train.columns
```

```python
train = train.drop(['Time_Orderd', 'Time_Order_picked'], axis = 1)
train.head()
```

```python
train = train[['Delivery_person_Age', 'Delivery_person_Ratings', 'Vehicle_condition',
       'multiple_deliveries', 'order_month', 'order_day',
       'diff_picked_min', 'Weatherconditions_conditions Fog',
       'Weatherconditions_conditions Sandstorms',
       'Weatherconditions_conditions Stormy',
       'Weatherconditions_conditions Sunny',
       'Weatherconditions_conditions Windy', 'Road_traffic_density_Jam ',
       'Road_traffic_density_Low ', 'Road_traffic_density_Medium ',
       'Type_of_order_Drinks ', 'Type_of_order_Meal ', 'Type_of_order_Snack ',
       'Type_of_vehicle_motorcycle ', 'Type_of_vehicle_scooter ',
       'Festival_Yes ', 'City_Semi-Urban ', 'City_Urban ', 'order_week_Monday',
       'order_week_Saturday', 'order_week_Sunday', 'order_week_Thursday',
       'order_week_Tuesday', 'order_week_Wednesday','Time_taken(min)']]
```

```python
X = train.iloc[:, :-1].values #처음부터 마지막 컬럼 직전까지의 데이터(독립변수)
y = train.iloc[:, -1].values #마지막 컬럼 데이터 (종속 변수 - 결과)
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression() #객체 생성
reg.fit(X, y)
```

```python
y_pred = reg.predict(X) #X에 대한 예측 값
y_pred
```

```python
plt.scatter(X, y, color = 'blue') #산점도
plt.plot(X, y_pred, color = 'green') #선 그래프
plt.title('delivery') #제목
plt.xlabel('X') #X 축 이름
plt.ylabel('y') #y 축 이름
plt.show()
```
